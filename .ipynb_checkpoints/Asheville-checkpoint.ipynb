```python
import pandas as pd
%matplotlib notebook
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt

filename = "data/listings.csv"
listings_data = pd.read_csv(filename)
listings_data.head()
```

```python
filename_reviews = "data/reviews_full.csv"
reviews_data = pd.read_csv(filename_reviews)
reviews_data.head()
```

```python
filename_calendar = "data/calendar.csv"
calendar_data = pd.read_csv(filename_calendar)
calendar_data.head()
```

```python
day_totals = calendar_data.groupby('listing_id').count()
day_totals.head()
```

```python
booked_out = calendar_data.loc[calendar_data["available"]=="f",:]
booked_out.head()
```

```python
available_false = calendar_data.loc[calendar_data["available"]=="f",:]
booked_dates = available_false.groupby('listing_id').count()['available']
booked_dates = pd.DataFrame(booked_dates)
booked_dates
```

```python
available_true = calendar_data.loc[calendar_data["available"]=="t",:]
nonbooked_dates = available_true.groupby('listing_id').count()['available']
nonbooked_dates = pd.DataFrame(nonbooked_dates)
nonbooked_dates
```
